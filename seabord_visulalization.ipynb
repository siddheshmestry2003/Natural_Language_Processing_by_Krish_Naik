### seaborn is a python visualization library based on matplotlib that proide a high level interface for drawing attractive and informative statistical fraphiical.seaboarn helps in creating complex vusualization with just a few lines of code .in this lesson ,we eill cover the basics of seaborn including creating various types of plotes and customizing them

```python
import seaborn as sns
```

```python
# basic ploting
tips=sns.load_dataset('tips')
tips
```

```python
# crate a scatter
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
print("scatter plot is total bill vs tip")
```

```python
# line plot

sns.lineplot(x='size',y='tip',data=tips)
pl

sns.lineplot(x='size',y='total_bill',data=tips)
```

```python
# categorical plot
#barplot

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('bar plot of total bijll by day')
```

```python
# box plot

sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# to show distribution we can use violinplot

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
# histogram
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
# histogram

sns.histplot(tips['total_bill'],bins=10)
```

```python


#kde plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# pairplot
sns.pairplot(data=tips)
```

```python
correlation=tips[['total_bill','tip','size']].corr()
correlation
```

```python
# heatmap
correlation=tips[['total_bill','tip','size']].corr()
sns.heatmap(correlation,annot=True)
```
